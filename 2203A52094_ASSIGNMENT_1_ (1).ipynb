```python
#Implement np.sin(), log(), log2(), np.exp()) function of an array
array1=np.array([10,20,30,40,50])
array2=np.sin(array1)
array3=np.cos(array1)
array4=np.tan(array1)
array5=np.log10(array1)
array6=np.log2(array1)
array7=np.exp(array1)
print("the sin array is",array2)
print("the cos array is",array3)
print("the tan array is",array4)
print("the log10 array is",array5)
print("the log base2 array is",array6)
print("the exp array is",array7)
```

```python
people = ['Ann','Chen','Imran','Lily','Sruthi']
age = [21,19,12,22,25]
weight = [55,43,77,34,87]
height = [160,234,173,246,105]
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python
Generate bar plot
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (n kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
